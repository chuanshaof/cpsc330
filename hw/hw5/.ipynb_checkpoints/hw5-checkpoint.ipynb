## 5. Preprocessing and transformations <a name="5"></a>
<hr>
rubric={points:10}

**Your tasks:**

1. Identify different feature types and the transformations you would apply on each feature type. 
2. Define a column transformer, if necessary. 

<div class="alert alert-warning">
    
Solution_5

```python
numeric_features = ["AGE", "LIMIT_BAL"]
minmax_features = ["PAY_1", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6"]
categorical_features = ["MARRIAGE"]
ordinal_features = ["EDUCATION"]
binary_features = ["SEX"]
passthrough_features = ['NEW_BILL_AMT1', 'NEW_BILL_AMT2', 'NEW_BILL_AMT3', 'NEW_BILL_AMT4', 'NEW_BILL_AMT5', 'NEW_BILL_AMT6', 
                        'NEW_PAY_AMT1', 'NEW_PAY_AMT2', 'NEW_PAY_AMT3', 'NEW_PAY_AMT4', 'NEW_PAY_AMT5', 'NEW_PAY_AMT6']
drop_features = ["ID",
                 "BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6", 
                 "PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]
```

```python
numeric_transformer = StandardScaler()

minmax_transformer = make_pipeline(SimpleImputer(strategy="mean"), MinMaxScaler())

categorical_transformer = make_pipeline(SimpleImputer(strategy='constant', fill_value="others"), OneHotEncoder(handle_unknown='ignore', sparse = False))

education_list = ["graduate school", "university", "high school", "others", "unknown"]
ordinal_transformer = make_pipeline(SimpleImputer(strategy="constant", fill_value="unknown"), OrdinalEncoder(categories=[education_list], dtype=int))

binary_transformer = OneHotEncoder(drop='if_binary', dtype=int)

preprocessor = make_column_transformer((numeric_transformer, numeric_features),
                                       (minmax_transformer, minmax_features),
                                       (ordinal_transformer, ordinal_features),
                                       (binary_transformer, binary_features),
                                       (categorical_transformer, categorical_features),
                                       ("passthrough", passthrough_features),
                                       ("drop", drop_features))

preprocessor
```

```python
"""
Transforming features here so that we don't have to do it so many times later
"""


X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)

column_names = (
    numeric_features 
    + minmax_features
    + ["education"]
    + preprocessor.named_transformers_["onehotencoder"].get_feature_names_out().tolist()
    + preprocessor.named_transformers_["pipeline-3"]["onehotencoder"].get_feature_names_out().tolist()
    + passthrough_features
)

display(pd.DataFrame(X_train_transformed, columns=column_names))

pd.DataFrame(X_train_transformed, columns=column_names).describe()
```

```python
# Taken from HW3

def mean_std_cross_val_scores(model, X_train, y_train, params=np.nan):
    """
    Returns mean and std of cross validation

    Parameters
    ----------
    model :
        scikit-learn model
    X_train : numpy array or pandas DataFrame
        X in the training data
    y_train :
        y in the training data

    Returns
    ----------
        pandas Series with mean scores from cross_validation
    """

    scoring = {'accuracy' : make_scorer(accuracy_score), 
               'precision' : make_scorer(precision_score, zero_division=False),
               'recall' : make_scorer(recall_score, zero_division=False), 
               'f1_score' : make_scorer(f1_score, zero_division=False)}


    scores = cross_validate(model, X_train, y_train, scoring=scoring, cv=5, return_train_score=True, error_score=np.nan)

    mean_scores = pd.DataFrame(scores).mean()
    std_scores = pd.DataFrame(scores).std()
    out_col = [params]

    for i in range(len(mean_scores)):
        out_col.append((f"%0.3f (+/- %0.3f)" % (mean_scores[i], std_scores[i])))

    return pd.Series(data=out_col, index=["Parameters"] + list(mean_scores.index))

results_dict = {}
```

```python
results_dict = {}

dummy = DummyClassifier(strategy='most_frequent')
results_dict["Dummy Classifier"] = mean_std_cross_val_scores(dummy, X_train, y_train)
results_df = pd.DataFrame(results_dict).T
results_df
```
